```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('log_week_2_end.csv')
df['duration_processed'] = df['duration'].apply(lambda x: float(str(x).replace('duration: ', '').replace(' ms', '')))
df['query_processed'] = df['query'].apply(lambda x: x.replace(',', ' ').split())
```

```python
df
```

```python
big_list = []
for list_ in df['query_processed'].to_list():
    big_list.extend(list_)
big_list
```

```python
processed_big_list = list(set(big_list))
total_length = len(processed_big_list)
```

```python
codes = pd.factorize(processed_big_list)
```

```python
codes[1][0]
```

```python
alphabet_df = pd.DataFrame(columns = ['code', 'value'])
for index in range(len(processed_big_list)):
    alphabet_df = alphabet_df.append({
        "code": codes[0][index],
        "value": codes[1][index]
    }, ignore_index=True)
```

```python
alphabet_df['code'] = alphabet_df['code'].apply(lambda x: x+1)
alphabet_df
```

```python
alphabet_df.to_csv('encoding.csv')
```
